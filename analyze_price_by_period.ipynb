# Import library and load dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
calendar = pd.read_csv('seattle/calendar.csv')
listings = pd.read_csv('seattle/listings.csv')
reviews = pd.read_csv('seattle/reviews.csv')
```

```python
listings.price.head();
listings_price = listings.price.replace('[\$,]', '', regex=True).astype(float)
listings_price.head()
```

```python
listings_price.hist(bins=100)
```

### Change date to datetime

```python
calendar.date = pd.to_datetime(calendar.date, format='%Y-%m-%d')
```

### Change price to float

```python
calendar['price'] = calendar.price.replace('[\$,]', '', regex=True).astype(float)
```

## Visualizing average price

```python
calendar_price_mean = calendar[['date', 'price']].groupby('date').mean()
calendar_price_mean.plot.line(figsize=(20,8))
```

```python
calendar_price_mean['date'] = calendar_price_mean.index
```

```python
calendar_price_mean[calendar_price_mean.date.map(lambda x: x.month) == 3].price.plot.line(figsize=(20,8))
```
