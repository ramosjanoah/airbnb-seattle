```python
calendar_groupby2 = calendar.groupby(['listing_id', 'date', 'available'])
```

```python
calendar_crosstab = pd.crosstab(calendar.listing_id, calendar.date, values=calendar.available, aggfunc='first')
```

```python
calendar_crosstab = calendar_crosstab.replace(to_replace='f', value=1).replace(to_replace='t', value=0)
```

```python
calendar_crosstab_T = calendar_crosstab.T
```

```python
calendar_crosstab_T.sum().sort_values().to_csv('test.csv')
```
